# 4.plot different kinds of graphs for mean value represetation ? (use required headers) ?

```python
auto.groupby('cylinders')['mpg'].mean().plot(kind='bar', color = 'peru');
plt.title("cylinders by City MPG")
plt.ylabel('mpg')
plt.xlabel('cylinders');
```

```python
auto.origin.value_counts().nlargest(10).plot(kind='bar', figsize=(10,5))
plt.title("Number of vehicles by origin")
plt.ylabel('Number of vehicles')
plt.xlabel('origin');
```

# 5.Draw the pair plot using sns for mpg, weight,orign and with hue origin, set the size to 4 

```python
sns.pairplot(auto,vars=["mpg","weight","origin"],hue='origin',size=4)
plt.show()
```
